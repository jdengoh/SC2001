```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

# Qn C(i)

## Empirical Analysis

```python
fixedS_data = pd.read_csv("../csv_files/fixed_S_results.csv")
fixedS_data.tail()
```

```python
y_worst = []
n_theory = []
y_best = []
for n in range(1000, 10000000, 499995):
    n_theory.append(n)
    y_theory.append(n*50 + n*np.log10(n/50))
    y_best.append(n+n*np.log10(n/50))
```

```python
x1 = fixedS_data['n']
y1 = fixedS_data['Avg Key Comp']
y2 = fixedS_data['Avg Processing Time']

#print(len(y1))

fig, axes = plt.subplots(1, 2, figsize=(16, 5))

axes[0].plot(x1, y1)
axes[0].plot(n_theory, y_theory)
axes[0].plot(n_theory, y_best)
axes[0].set_title('Average number of key comparisons vs n value')
axes[0].set_xlabel('Values of n')
axes[0].set_ylabel('Number of key comparisons')

axes[1].plot(x1, y2)
axes[1].set_title('Average Processing Time vs n value')
axes[1].set_xlabel('Values of n')
axes[1].set_ylabel('Average Processing Time')
```

## Theoretical Analysis

```python
#fixedS
N = np.array(range(1, 1000000))
S = 50
y = N*S + N*(np.log(N/S))
plt.plot(N,y)
plt.xlabel('Size of Array, N')
plt.title('Theoretical Time Complexity for Hybrid Sort')
plt.grid(True)
plt.show()
```

# Qn C(iii)

```python
data_1k = pd.read_csv("../csv_files/optimalS_1000_results.csv")
data_10k = pd.read_csv("../csv_files/optimalS_10000_results.csv")
data_100k = pd.read_csv("../csv_files/optimalS_100000_results.csv")
data_1mil = pd.read_csv("../csv_files/optimalS_1000000_results.csv")
data_10mil = pd.read_csv("../csv_files/optimalS_10000000_results.csv")
```

```python
x = data_1k['S']

comp_1k = data_1k['Avg Key Comp']
time_1k = data_1k['Avg Processing Time']
comp_10k = data_10k['Avg Key Comp']
time_10k = data_10k['Avg Processing Time']
comp_100k = data_100k['Avg Key Comp']
time_100k = data_100k['Avg Processing Time']
comp_1mil = data_1mil['Avg Key Comp']
time_1mil = data_1mil['Avg Processing Time']
comp_10mil = data_10mil['Avg Key Comp']
time_10mil = data_10mil['Avg Processing Time']
```

```python
fig, axes = plt.subplots(5, 2, figsize=(10, 20))

axes[0,0].plot(x, comp_1k)
axes[0,0].set_title('Average number of key comparisons vs S')
axes[0,0].set_xlabel('Values of S')
axes[0,0].set_ylabel('Average number of key comparisons')

axes[0,1].plot(x, time_1k)
axes[0,1].set_title('Average Processing Time vs S')
axes[0,1].set_xlabel('Values of S')
axes[0,1].set_ylabel('Average Processing Time')

axes[1,0].plot(x, comp_10k)
axes[1,0].set_title('Average number of key comparisons vs S')
axes[1,0].set_xlabel('Values of S')
axes[1,0].set_ylabel('Average number of key comparisons')

axes[1,1].plot(x, time_10k)
axes[1,1].set_title('Average Processing Time vs S')
axes[1,1].set_xlabel('Values of S')
axes[1,1].set_ylabel('Average Processing Time')

axes[2,0].plot(x, comp_100k)
axes[2,0].set_title('Average number of key comparisons vs S')
axes[2,0].set_xlabel('Values of S')
axes[2,0].set_ylabel('Average number of key comparisons')

axes[2,1].plot(x, time_100k)
axes[2,1].set_title('Average Processing Time vs S')
axes[2,1].set_xlabel('Values of S')
axes[2,1].set_ylabel('Average Processing Time')

axes[3,0].plot(x, comp_1mil)
axes[3,0].set_title('Average number of key comparisons vs S')
axes[3,0].set_xlabel('Values of S')
axes[3,0].set_ylabel('Average number of key comparisons')

axes[3,1].plot(x, time_1mil)
axes[3,1].set_title('Average Processing Time vs S')
axes[3,1].set_xlabel('Values of S')
axes[3,1].set_ylabel('Average Processing Time')

axes[4,0].plot(x, comp_10mil)
axes[4,0].set_title('Average number of key comparisons vs S')
axes[4,0].set_xlabel('Values of S')
axes[4,0].set_ylabel('Average number of key comparisons')

axes[4,1].plot(x, time_10mil)
axes[4,1].set_title('Average Processing Time vs S')
axes[4,1].set_xlabel('Values of S')
axes[4,1].set_ylabel('Average Processing Time')

plt.tight_layout()
```
