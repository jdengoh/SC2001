```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

# Qn C(i)

## Empirical Analysis

```python
fixedS_data = pd.read_csv("../csv_files/fixed_S_results.csv")
fixedS_data.tail()
```

```python
y_worst = []
n_theory = []
y_best = []
for n in range(1000, 10000000, 499995):
    n_theory.append(n)
    y_worst.append(n*50 + n*np.log10(n/50))
    y_best.append(n+n*np.log10(n/50))
```

```python
x1 = fixedS_data['n']
y1 = fixedS_data['Avg Key Comp']
y2 = fixedS_data['Avg Processing Time']

#print(len(y1))

fig, axes = plt.subplots(1, 2, figsize=(16, 5))

axes[0].plot(x1, y1, label = "Empirical")
axes[0].plot(n_theory, y_worst, label = "Worst case")
axes[0].plot(n_theory, y_best, label = "Best case")
axes[0].legend()

axes[0].set_title('Average number of key comparisons vs n value')
axes[0].set_xlabel('Values of n')
axes[0].set_ylabel('Number of key comparisons')

axes[1].plot(x1, y2)
axes[1].set_title('Average Processing Time vs n value')
axes[1].set_xlabel('Values of n')
axes[1].set_ylabel('Average Processing Time')
```

## Theoretical Analysis

```python
#fixedS
N = np.array(range(1, 1000000))
S = 50
y = N*S + N*(np.log(N/S))
plt.plot(N,y)
plt.xlabel('Size of Array, N')
plt.title('Theoretical Time Complexity for Hybrid Sort')
plt.grid(True)
plt.show()
```

# Qn C(ii)

## Empirical Analysis

```python
fixedn_data = pd.read_csv("../csv_files/fixed_n_results.csv")
fixedn_data.head()
```

```python
y_worst = []
y_best = []
x_theory = []
for S in range(1, 51, 1):
    x_theory.append(S)
    y_worst.append(1000000*S + 1000000*np.log10(1000000/S))
    y_best.append(1000000 + 1000000*np.log10(1000000/S))
```

```python
x1 = fixedn_data['S']
y1 = fixedn_data['Avg Key Comp']
y2 = fixedn_data['Avg Processing Time']

#print(len(y1))

fig, axes = plt.subplots(1, 2, figsize=(16, 5))


axes[0].plot(x1, y1, label = "Empirical")
axes[0].plot(x_theory, y_worst, label = "Worst Case")
axes[0].plot(x_theory, y_best, label = "Best Case")
axes[0].set_title('Average number of key comparisons vs S value')
axes[0].set_xlabel('Values of S')
axes[0].set_ylabel('Number of key comparisons')
axes[0].legend()

axes[1].plot(x1, y2)
axes[1].set_title('Average Processing Time vs S value')
axes[1].set_xlabel('Values of S')
axes[1].set_ylabel('Average Processing Time')
```

## Theoretical Analysis

```python
#fixedN
S = np.array(range(1, 100))
N = 1000000
y = N*S + N*(np.log(N/S))
plt.plot(S,y)
plt.xlabel('value of S, S')
plt.ylabel('Time Complexity')
plt.title('Theoretical Time Complexity for Hybrid Sort ')
plt.grid(True)
plt.show()
```

# Qn D

```python
merge_data = pd.read_csv("../csv_files/Merge_results.csv")
mergeInsert_data = pd.read_csv("../csv_files/MergeInsert_results.csv")

display(merge_data)
display(mergeInsert_data)
```

```python
# Concatenate 2 dataframes together
algo = pd.concat([merge_data, mergeInsert_data], axis = 0, ignore_index = True)
method = ["Merge", "Hybrid"]
algo.insert(0, "Algorithm", method)
display(algo)
```

```python
# Plot a bar chart to visualise the comparison

ax = sb.barplot(data = algo, x = algo['Algorithm'], y = algo['Avg Key Comp'])
ax.set_title('Merge vs Hybrid (Key Comparison)')
ax.bar_label(ax.containers[0])

plt.show()

ax = sb.barplot(data = algo, x = algo['Algorithm'], y = algo['Avg Processing Time'])
ax.set_title('Merge vs Hybrid (Processing Time)')
ax.bar_label(ax.containers[0])
ax.set_ylabel('Average Processing Time/s')
plt.show()
```
