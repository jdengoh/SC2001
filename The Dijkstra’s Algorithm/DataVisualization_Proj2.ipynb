```python
varying_V = pd.read_csv("varying_V.txt", delimiter=",")

varying_V.head()
```

```python
v = varying_V["V"]
e = varying_V["E"]

max_empirical = max(varying_V["Min_Heap Queue Runtime"])
max_theoretical = max(((v+e)*np.log2(v))*3.0)  # Adjusted for V^3*logV complexity
constant = max_empirical / max_theoretical

alist = pd.DataFrame([v, (v+e)*np.log2(v)*3.0*constant]).transpose()
#alist = pd.DataFrame([v, v*e*np.log2(v)*1100]).transpose()
array = pd.DataFrame([v, v*v*3.1]).transpose()
array2 = pd.DataFrame([v, v*v*4]).transpose()
alist.columns = ["V", "runtime"]
array.columns = ["V", "runtime"]
array2.columns = ["V", "runtime"]
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 15))
plt.plot(array2["V"], array2["runtime"])
plt.plot(varying_V["V"], varying_V["Array Queue Runtime"])
plt.plot(array["V"], array["runtime"])

plt.legend(["theoretical (4*V^2)", "empirical", "theoretical (3.1*V^2)"], loc ="lower right")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 15))
plt.plot(varying_V["V"], varying_V["Min_Heap Queue Runtime"])
plt.plot(alist["V"], alist["runtime"])
plt.legend(["empirical", "theoretical (V*E*logV"], loc ="lower right")
```

```python
fig, ax1 = plt.subplots(figsize=(10, 5))


# Plot Array Queue Runtime on primary y-axis
ax1.plot(small_E["V"], small_E["Array Queue Runtime"], 'b-')
ax1.set_xlabel('Number of Nodes, V')
ax1.set_ylabel('Array Queue Runtime', color='b')
ax1.tick_params('y', colors='b')
ax1.legend(["Array"], loc="upper left")

# Create a secondary y-axis for Min_Heap Queue Runtime
ax2 = ax1.twinx()
ax2.plot(small_E["V"], small_E["Min_Heap Queue Runtime"], 'r-')
ax2.set_ylabel('Min_Heap Queue Runtime', color='r')
ax2.tick_params('y', colors='r')
ax2.legend(["AdjList"], loc="upper right")

plt.show()
```
